```python
import time # 時刻DBへのアクセス
import sys # システムのライブラリ
import requests # HTTPリクエスト用ライブラリ
import re # 正規表現
from bs4 import BeautifulSoup # HTMLパーサ
```

```python
url = 'http://itpro.nikkeibp.co.jp/article/COLUMN/20000518/200024/?TOC='
```

```python
for nCnt in range(1,56):
    print("[" + str(nCnt) + "]")
    time.sleep(1)
    res = requests.get(url + str(nCnt))

    soup = BeautifulSoup(res.content,'html.parser')

    records = soup.select('a')
    for record in records:
        if re.match('^<h3>',str(record.parent)):
            if re.search('/article/|/members/',str(record)):
                print(str(record))
```
